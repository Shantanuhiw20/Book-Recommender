##Popularity based recommendation system##

```python
ratings_with_name = ratings.merge(books, on='ISBN')
```

```python
ratings_with_name
```

```python
num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
```

```python
num_rating_df.rename(columns={'Book-Rating': 'num_ratings'}, inplace=True)
```

```python
num_rating_df
```

```python
avg_rating_df = ratings_with_name.groupby('Book-Title').mean()['Book-Rating'].round(3).reset_index()
avg_rating_df.rename(columns={'Book-Rating': 'avg_ratings'}, inplace=True)
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df, on='Book-Title')
popular_df 
```

```python
popular_df = popular_df[popular_df['num_ratings']>=250].sort_values('avg_ratings', ascending=False).head(50)
```

```python
popular_df.merge(books, on='Book-Title').drop_duplicates('Book-Title').shape
```

```python
print(popular_df.columns)
```

```python
popular_df = popular_df.merge(books, on='Book-Title').drop_duplicates('Book-Title')[['Book-Title', 'Book-Author', 'Image-URL-M', 'num_ratings', 'avg_ratings']]
```

**Collaborating Based Filtering**

Criteria : 
1) Users who have rated more than 200 times
2) Books which have minimum 50 ratings

```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200
```

```python
x[x]
```

```python
padhe_likhe_users = x[x].index
```

```python
ratings_with_name['User-ID'].isin(padhe_likhe_users)
```

```python
ratings_with_name[ratings_with_name['User-ID'].isin(padhe_likhe_users)]
```

```python
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(padhe_likhe_users)]
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating'] >= 50
```

```python
y[y] #Boolean indexing
```

```python
famous_books = y[y].index
```

```python
famous_books
```

```python
filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_ratings.drop_duplicates()
```

```python
pt = final_ratings.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating')
```

```python
pt.fillna(0, inplace=True)
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
```

```python
similarity_score
```

```python
def recommend(book_name):
    index = np.where(pt.index == book_name)[0][0]
    similar_items = list(enumerate(similarity_score[index])) #enumerate gives us the index and the value of the item
    sorted_similar_items = sorted(similar_items, key=lambda x: x[1], reverse=True)[1:8] #Similar Items
    recommended_books = []
    for item in sorted_similar_items:
        recommended_books.append(pt.index[item[0]])
    return recommended_books
```

```python
def recommend1(book_name):
    #index fetch
    index = np.where(pt.index == book_name)[0][0]
    #similarity score fetch
    similar_items = sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:8] #enumerate gives us the index and the value of the item
    #sort the items based on similarity score

    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))

        data.append(item)
    
    return data
```
